```python
from pathlib import Path
from docling.document_converter import DocumentConverter
```

```python
# data path
cwd = Path.cwd()

def find_project_root(start: Path) -> Path:
    for parent in [start, *start.parents]:
        if (parent / "pyproject.toml").exists() or (parent / ".git").exists():
            return parent
    return start

project_root = find_project_root(cwd)
print("cwd:", cwd)
print("project_root:", project_root)

data_path = project_root / "data" / "MAS Notice 626.pdf"
```

```python
from docling.document_converter import DocumentConverter

pdf_path = r"C:\Users\yq198\Desktop\DANNY_AI\LLM_Compliance\data\MAS Notice 626.pdf"

converter = DocumentConverter()

all_docs = []

# process in chunks of 10 pages (safer than 20)
for start in range(1, 100, 10):  # adjust 100 → total pages
    end = start + 9
    print(f"Processing pages {start}-{end}")

    try:
        result = converter.convert(
            pdf_path,
            page_range=(start, end)
        )
        all_docs.append(result.document)

    except Exception as e:
        print(f"Failed pages {start}-{end}: {e}")
```

```python
print(result.document.export_to_markdown()) 
md_text = result.document.export_to_markdown()

with open(r"C:\Users\yq198\Desktop\DANNY_AI\LLM_Compliance\data\mas_626.md", "w", encoding="utf-8") as f:
    f.write(md_text)
```
